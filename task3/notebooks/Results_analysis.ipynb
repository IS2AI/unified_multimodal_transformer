### Libraries

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

### RGB

```python
resnet1_1 = torch.load("/workdir/results/resnet1_1_logs")
resnet1_2 = torch.load("/workdir/results/resnet1_2_logs")
resnet2_1 = torch.load("/workdir/results/resnet2_1_logs")
resnet2_2 = torch.load("/workdir/results/resnet2_2_logs")
vit1_1 = torch.load("/workdir/results/vit1_1_logs")
vit1_2 = torch.load("/workdir/results/vit1_2_logs")
```

```python
print(f"Best EER ViT L2:{vit1_1['best_eer']}")
print(f"Best EER ViT cos sim:{vit1_2['best_eer']}")
print(f"Best EER ResNet timm L2:{resnet2_1['best_eer']}")
print(f"Best EER ResNet timm cos sim:{resnet2_2['best_eer']}")
```

```python
rows = 2 # L2 and cos
columns = 4 # 'train_loss', 'train_acc', 'val_eer', 'val_acc'
fig = plt.figure(figsize=(5*columns, 4*rows))
grid = plt.GridSpec(rows, columns, wspace = .25, hspace = .25,figure=fig)
fig.suptitle("RGB", fontsize="x-large")

for cols, mode in enumerate(list(resnet1_1.keys())[:4]):

    row = 0 # L2
        
    exec (f"plt.subplot(grid{[row * columns + cols]})")

    plt.plot(np.arange(len(resnet1_1[mode])), resnet1_1[mode], label='resnet1, L2')
    plt.plot(np.arange(len(resnet2_1[mode])), resnet2_1[mode], label='resnet2, L2')
    plt.plot(np.arange(len(vit1_1[mode])), vit1_1[mode], label='vit1, L2')
    plt.title(f"{mode}")
    plt.xlabel("Epoch")
    plt.ylabel(f"{mode}")
    plt.legend()

    row = 1 # Cos Sim
    exec (f"plt.subplot(grid{[row * columns + cols]})")
    plt.plot(np.arange(len(resnet1_2[mode])), resnet1_2[mode], label='resnet1, cos sim')
    plt.plot(np.arange(len(resnet2_2[mode])), resnet2_2[mode], label='resnet2, cos sim')
    plt.plot(np.arange(len(vit1_2[mode])), vit1_2[mode], label='vit1, cos sim')
    plt.xlabel("Epoch")
    plt.ylabel(f"{mode}")
    plt.title(f"{mode}")
    plt.legend()
```

```python
m = torch.load("/workdir/results/vit1_1_logs")
plt.plot(np.arange(len(m["train_loss"])), m["train_loss"],c='b',label="train_loss")
plt.plot(np.arange(len(m["train_acc"])), m["train_acc"],c='g',label="train_acc")
plt.plot(np.arange(len(m["val_eer"])), m["val_eer"],c='r',label="val_eer")
plt.plot(np.arange(len(m["val_acc"])), m["val_acc"],c='y',label="val_acc")
plt.legend()
```

```python
m = torch.load("/workdir/results/vit1_2_logs")
plt.plot(np.arange(len(m["train_loss"])), m["train_loss"],c='b',label="train_loss")
plt.plot(np.arange(len(m["train_acc"])), m["train_acc"],c='g',label="train_acc")
plt.plot(np.arange(len(m["val_eer"])), m["val_eer"],c='r',label="val_eer")
plt.plot(np.arange(len(m["val_acc"])), m["val_acc"],c='y',label="val_acc")
plt.legend()
```
